```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import MinMaxScaler
from sklearn.svm import SVC
```

```python
# data = pd.read_csv('./Data/Training_dataset_Small.csv')
data = pd.read_csv('training_data_v2.csv', low_memory=False)
```

```python
data.head()
```

```python
trainDataKey = data['application_key']
data.drop('application_key', axis=1, inplace=True)
```

```python
X = data[data.loc[:, data.columns != 'default_ind'].columns]
Y = data['default_ind']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=.2, random_state=1)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
print(X_train.head())
```

```python
clf = SVC()
clf = SVC(kernel='rbf', gamma=0.001, C=100)
print("Done")
```

```python
clf.fit(X_train, Y_train)
print("Done")
```

```python
print("Yes")
```

```python
scores = cross_val_score(clf, X_train, Y_train, cv=10)
print('CV accuracy: %.3f +/- %.3f' % (np.mean(scores), np.std(scores)))
```

```python
scores = cross_val_score(clf, X_test, Y_test, cv=10)
print('CV accuracy: %.3f +/- %.3f' % (np.mean(scores), np.std(scores)))
```
